```python
# takes a dataset, startrange and endrange define the Probability interval
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot # for plotting
    from scipy.stats import norm # normal distribution tool
    import seaborn as sns # plotting library 
    
    # draw histogram of dataset in green
    # kernal density establishment curve in blue 
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    
    # axvline - indicate vertical line
    # we are drawing 2 vertical line to define startrange and endrange in red 
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    
    # generate a sample
    sample = dataset
    
    # calculate parameters like mean and standerd deviation
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    
    # create a normal distribution object using mean and std 
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    
    # generate every integer b/w startrange and endrange
    # computes the pdf (probability dendity) value at each integer
    # sum those density as rough approximation of the area under the curve (prob mass) in that range 
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
    
```

```python
# one liner for loop
def get_range_probability(dist, startrange, endrange):
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]
    prob = sum(probabilities)
    print("The area between range({},{}):{}".format(startrange, endrange, prob))
    return prob
```
